```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data.drop('Status',axis=1),
train_data.Status, test_size=0.2, random_state=42)
```

```python
print('X_train :',X_train.shape)
print('X_test  :',X_test.shape)
print('y_train :',y_train.shape)
print('y_test  :',y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logit = LogisticRegression(multi_class='multinomial', solver='saga',max_iter=5000,class_weight='balanced')
```

```python
X_train
```

```python
test_data.isnull().sum()
```
